```python
batting = df2[['yearID', 'HR']]

batting_50_years = batting[batting['yearID'] > 1969]

batting_50_years.groupby('yearID').sum()

```

```python
%matplotlib inline
plt.plot(batting_50_years.groupby('yearID').sum(), color = 'red')
plt.title('Total Home Runs per Year')
plt.xlabel('Year')
plt.ylabel('Total Home Runs')
plt.show()
```

The above graph shows total home runs hit by Major League Baseball players over the last 50 years. Notice the general upward trend. Also, notice the spike in the late 90's/early 2000's (the "Steroid Era"), subsequent dip and then another spike. Are the balls "juiced"?

```python
scatter_data = df2[['yearID', 'teamID', 'H', 'R' ]]
sox = scatter_data[scatter_data['teamID']== 'CHA']
sox_1900 = sox[sox['yearID'] > 1900]
sox_data = sox_1900.groupby('yearID').sum()

```

```python
cubs = scatter_data[scatter_data['teamID']== 'CHN']
cubs_1900 = cubs[cubs['yearID'] > 1900]
cubs_data = cubs_1900.groupby('yearID').sum()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.scatter(x=sox_data['H'], y=sox_data['R'], color='black')
plt.ylabel('Runs')
plt.xlabel('Hits')
plt.xlim([800, 1800])
plt.ylim([200, 1000])
plt.title('White Sox Hits vs Runs')

plt.subplot(1,2,2)
plt.scatter(x=cubs_data['H'], y=cubs_data['R'], color='blue')
plt.ylabel('Runs')
plt.xlabel('Hits')
plt.xlim([800, 1800])
plt.ylim([200, 1000])
plt.title('Cubs Hits vs Runs')
plt.show()
```

The above subplot shows two teams - the White Sox and Cubs. Each scatter plot shows the respective team's runs vs hits every year since 1901. As expected, the scatter plots tell us that hits effect runs shown by the positive linear relationship between the variables. Comparing the two teams, the plots tell us they have fairly similar hit to run conversion.

```python
df4 = pd.read_csv('Fielding.csv')
```

```python
bar_data = df4[['yearID', 'teamID', 'E']]
sox_fielding = bar_data[bar_data['teamID']== 'CHA']
sox_fielding_1997 = sox_fielding[sox_fielding['yearID'] > 1996]
sox_data_fielding = sox_fielding_1997.groupby('yearID').sum()
sox_data_fielding
```

```python
years = sox_data_fielding.index
y_pos = np.arange(len(years))
errors = sox_data_fielding['E']

plt.bar(years, errors)
plt.xticks(years, rotation='vertical')
plt.show()
```

The above chart is a simple bar graph showing White Sox team errors from 1997 through 2018. The graph doesn't show much of a trend but over recent years, errors have been on an upward climb after three years of declining team errors starting in 2013.
