```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```
#upload the dataset
ufc = pd.read_csv('https://raw.githubusercontent.com/kylemcq13/dataset/master/ufc_data.csv')
```

```
#preview the dataset
ufc.head()
```

```
#list of column names
ufc.columns.values
```

```
#present all the data
ufc
```

```
ufc2 = ufc.plot.line(x = 'weight_class', y='R_Weight_lbs')

plt.title('Line Plot')
plt.ylabel('Weight lbs')
plt.xlabel('weight_class')

plt.show()
```

```
ufc3 = ufc.plot.bar(x = 'weight_class', y='R_Weight_lbs')

plt.title('Line Plot')
plt.ylabel('Weight lbs')
plt.xlabel('weight_class')

plt.show()
```
